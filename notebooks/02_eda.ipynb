```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
dataframe  = pd.read_csv('../data/cleaned_data.csv')
sns.histplot(dataframe['Price'], bins=50)
plt.title('Price Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(np.log1p(dataframe['Price']), bins=50)
plt.title('Log Price Distribution')
plt.xlabel('Log Price')
plt.show()
```

```python
plt.figure(figsize=(8,6))
numeric_cols = ['Year', 'Kilometers_Driven', 'Transmission', 'Owner_Type', 'Price']
sns.heatmap(dataframe[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='Year', y='Price', data=dataframe, alpha=0.3)
plt.title('Price vs Year')
plt.show()
```

```python
# 4. Price vs Kilometers Driven
plt.figure(figsize=(10,5))
sns.scatterplot(x='Kilometers_Driven', y='Price', data=dataframe, alpha=0.3)
plt.title('Price vs Kilometers Driven')
plt.show()
```

```python
# Average price by brand - need to reconstruct brand from one hot columns
# Price by transmission boxplot
plt.figure(figsize=(8,5))
sns.boxplot(x='Transmission', y='Price', data=dataframe)
plt.title('Price by Transmission')
plt.xticks([0,1], ['Automatic', 'Manual'])
plt.show()
```
